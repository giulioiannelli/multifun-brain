```python
'''#print una matrice di correlazione
trym = Diz_corr_features[keys13[0] + "_" + names_channels[0]]
fig = plt.figure(figsize=(10, 5))
plt.imshow(trym, interpolation='nearest')
plt.colorbar()'''
#ok

import bicm
print(bicm.__version__)

from bicm import BipartiteGraph
import bicm.models_functions as mof
import bicm.solver_functions as sof
import bicm.network_functions as nef
from tqdm import tqdm
import bicm.poibin as pb


def calculate_fdr_threshold(provapvmat, intervallo):
    D = provapvmat.shape[0] * (provapvmat.shape[0] - 1) / 2
    #diverso qui, ho una matrice bipartita non simmetrica e non quadrata
    #D = provapvmat.shape[0] * provapvmat.shape[1]
    sorted_indices = []
    sortedpvaluesfdr = []
    
    for i in range(provapvmat.shape[0]):
        for j in range(i + 1, provapvmat.shape[0]): #anche qui cambio dato che ho bipartita non simmetrica e non quadrata
        #for j in range(provapvmat.shape[1]):
            sortedpvaluesfdr.append(provapvmat[i][j])
            sorted_indices.append((i, j))
    
    sorted_pairs = sorted(zip(sortedpvaluesfdr, sorted_indices))  # Ordinamento congiunto
    sortedpvaluesfdr, sorted_indices = zip(*sorted_pairs)

    if len(sortedpvaluesfdr) == 0:
        print("Nessun valore soddisfa la condizione")
        return [], [], None

    sortedpvaluesfdr = np.array(sortedpvaluesfdr)
    thresholds = np.arange(1, len(sortedpvaluesfdr) + 1) * intervallo / D
    valid_indices = np.where(sortedpvaluesfdr <= thresholds)[0]

    if len(valid_indices) == 0:
        print("Nessun valore soddisfa la condizione")
        return [], [], None

    sogliapos = valid_indices[-1]
    soglia = (sogliapos + 1) * intervallo / D
    
    positionvalidate = []
    pvvalidati = []
    
    for i in range(len(sortedpvaluesfdr)):
        if sortedpvaluesfdr[i] <= soglia:
            positionvalidate.append(sorted_indices[i])
            pvvalidati.append(sortedpvaluesfdr[i])
        else:
            break
    
    if soglia is None:
        soglia = 0

    return positionvalidate, pvvalidati, soglia
```
